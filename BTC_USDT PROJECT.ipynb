```python
def calculate_atr(high, low, close, lookback=14):

    tr = np.maximum(high - low, 
                    np.maximum(abs(high - close.shift(1)), 
                              abs(low - close.shift(1))))
    
    atr = np.zeros_like(close)
    atr[lookback:] = np.mean(tr[-lookback:], axis=0)

    return atr
```

```python
def calculate_mean_reversion_profit(prices, entry_price, exit_price):
    

    if exit_price > entry_price:
        return (exit_price - entry_price) / entry_price * 100
    else:
        return (entry_price - exit_price) / entry_price * 100
```

```python
def calculate_trend_following_profit(prices, entry_price, exit_price):
    
    trend_direction = 1 if prices[-1] > prices[0] else -1

    if trend_direction * exit_price > trend_direction * entry_price:
        return (exit_price - entry_price) / entry_price * 100
    else:
        return (entry_price - exit_price) / entry_price * 100
```
